```r
ff = read.csv("~/Desktop/amazondata2.csv")
head(ff)
```

```r
# remove the '$' from Item.Total column
ff$Item.Total<-sub("[[:punct:]]","",ff$Item.Total)
head(ff$Item.Total,3)
```

```r
#Check class of Item.Total 
class(ff$Item.Total)
```

```r
#Change class to numeric so operations can be used
ff$Item.Total <- as.numeric(ff$Item.Total)
head(ff$Item.Total,3)
```

```r
# basic statistic analysis of Item Total Date
sum(ff$Item.Total)
min(ff$Item.Total)
max(ff$Item.Total)
median(ff$Item.Total)
mean(ff$Item.Total)
sd(ff$Item.Total)
```

```r
#Remove '$' from Tax column
ff$Item.Subtotal.Tax<-sub("[[:punct:]]","",ff$Item.Subtotal.Tax)
head(ff$Item.Subtotal.Tax,3)
```

```r
#Check class of tax column
class(ff$Item.Subtotal.Tax)
```

```r
#Change class to numeric so operations can be used
ff$Item.Subtotal.Tax <- as.numeric(ff$Item.Subtotal.Tax)
```

```r
sum(ff$Item.Subtotal.Tax)
mean(ff$Item.Subtotal.Tax)
min(ff$Item.Subtotal.Tax)
max(ff$Item.Subtotal.Tax)
median(ff$Item.Subtotal.Tax)
```

```r

# find the tax rate
sum(ff$Item.Subtotal.Tax) / sum(ff$Item.Total)
```

```r
#Changes '/' to '-' so I can change to y-m-d format and checks class
ff$Order.Date<-gsub("[[:punct:]]","-",ff$Order.Date)
head(ff$Order.Date)
class(ff$Order.Date)
```

```r
#Changes class to date
ff$Order.Date <- as.Date(ff$Order.Date,
                        format = "%m-%d-%y")
head(ff$Order.Date)
class(ff$Order.Date)
```
